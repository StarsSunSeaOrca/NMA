```python
#@title Data retrieval
import os, requests

fname = "stringer_orientations.npy"
url = "https://osf.io/ny4ut/download"

if not os.path.isfile(fname):
  try:
    r = requests.get(url)
  except requests.ConnectionError:
    print("!!! Failed to download data !!!")
  else:
    if r.status_code != requests.codes.ok:
      print("!!! Failed to download data !!!")
    else:
      with open(fname, "wb") as fid:
        fid.write(r.content)
```

```python
from IPython.testing import IPYTHON_TESTING_TIMEOUT_SCALE
def orientation_degree(data_min, data_max, value):
  # Number of bins
  num_bins = 360

  # Calculate the range of values and bin width
  bin_width = (data_max - data_min) / num_bins

  degree = int((value - data_min) // bin_width)
  return degree

data= dat['istim']
data_min = np.min(data)
data_max = np.max(data)

# test data
test_data = np.array([[1,2.62022653e+00], [2,4.52593227e+00]])

# convert np array to python list
list_from_arr = test_data.tolist()

for neuron in list_from_arr:
  istim = neuron[1]
  degree = orientation_degree(data_min, data_max, istim)
  neuron.append(degree)

# convert python list to np array
neuron_istim_degree = np.array(list_from_arr)
print(neuron_istim_degree)
```
